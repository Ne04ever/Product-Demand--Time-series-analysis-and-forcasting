```python
### Lets group the data wih respect to each product catagory
furniture_df = df[df['Category']=='Furniture']
office_df = df[df['Category']=='Office Supplies']
tech_df = df[df['Category']=='Technology']
```

```python
#taking only the necessary columns and sorting
furniture_df = furniture_df[['Order Date','Sales']]
furniture_df = furniture_df.sort_values('Order Date')
furniture_df = furniture_df.groupby('Order Date')['Sales'].sum().reset_index()
#setting order date as index
furniture = furniture_df.set_index('Order Date')
```

```python
#taking only the necessary columns and sorting
office_df = office_df[['Order Date','Sales']]
office_df = office_df.sort_values('Order Date')
office_df = office_df.groupby('Order Date')['Sales'].sum().reset_index()
#setting order date as index
office = office_df.set_index('Order Date')
```

```python
#taking only the necessary columns and sorting
tech_df = tech_df[['Order Date','Sales']]
tech_df = tech_df.sort_values('Order Date')
tech_df = tech_df.groupby('Order Date')['Sales'].sum().reset_index()
#setting order date as index
tech = tech_df.set_index('Order Date')
```

```python
# Saving the dataframe
```

```python
furniture.to_csv('furniture.csv')
office.to_csv('office.csv')
tech.to_csv('tech.csv')
```

```python
# Anlysing the data
dic_total = {'furniture':furniture,
           'office':office,
           'tech': tech}
df_total = pd.concat(dic_total, axis= 1)
plt.style.use('fivethirtyeight')
df_total.plot(figsize = (20,6))
```

```python
#resambling
m_furniture = furniture['Sales'].resample('M').sum()
m_tech = tech['Sales'].resample('M').sum()
m_office = office['Sales'].resample('M').sum()
```

```python
m_furniture.plot(figsize=(20,6))
plt.title('Furniture Sale')
```

```python
m_office.plot(figsize=(20,6))
plt.title('office Sale')
```

```python
m_tech.plot(figsize=(20,6))
plt.title('tech Sale')
```

```python
dic2_total = {'furniture':m_furniture,
           'office':m_office,
           'tech':m_tech}
df_total2 = pd.concat(dic2_total, axis= 1)
df_total2.plot(figsize = (20,6))
```
